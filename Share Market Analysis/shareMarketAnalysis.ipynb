```python
import findspark
findspark.init()
```

```python
from pyspark.sql import SparkSession
```

```python
from pyspark.sql.functions import *
```

```python
spark = SparkSession.builder.master("local[1]").appName("shareMarketAnalysis").getOrCreate()    
```

```python
rdd = spark.sparkContext.textFile("sharemarket.csv")
```

```python
rdd1 = rdd.map(lambda x : x.split(","))
```

```python
rdd2 = rdd1.map(lambda x : x[0:11] + x[12:15])
```

```python
rdd2.collect()
```

```python
df = rdd2.toDF(schema = ['MARKET', 'SERIES', 'SYMBOL', 'SECURITY', 'PREV_CL_PR', 'OPEN_PRICE', 'HIGH_PRICE', 'LOW_PRICE', 'CLOSE_PRICE', 
                         'NET_TRDVAL', 'NET_TRDQTY', 'TRADES', 'HI_52_WK', 'LO_52_WK'])
```

```python
df.show()
```

```python
df.createOrReplaceTempView("sharemarket")
```

```python
#1
count_series = spark.sql("select count(distinct(SERIES)) as count from sharemarket")
```

```python
count_series.show()
```

```python
#2
display_data = spark.sql("select distinct(SERIES) from sharemarket")
```

```python
display_data.show()
```

```python
#3
sum_prices = spark.sql("select series, sum(PREV_CL_PR), sum(OPEN_PRICE), sum(HIGH_PRICE), sum(LOW_PRICE), sum(CLOSE_PRICE), \
                 sum(NET_TRDVAL), sum(HI_52_WK), sum(LO_52_WK) from sharemarket GROUP BY series")
```

```python
sum_prices.show(vertical = True)
```

```python
df = df.withColumn("NET_TRDVAL", col("NET_TRDVAL").cast("float"))
df = df.withColumn("NET_TRDQTY", col("NET_TRDQTY").cast("float"))
df = df.withColumn("TRADES", col("TRADES").cast("integer"))
```

```python
#4
highest_net_trade = df.groupBy(['series', 'security']).sum('NET_TRDVAL').alias('max_net_trade')
```

```python
highest_net_trade.show()
```

```python
highest_net_trade = highest_net_trade.sort(desc("sum(NET_TRDVAL)"))
```

```python
highest_net_trade.show()
```

```python
highest_net_trade.first()
```

```python
#5
df1 = df.withColumn("ALL_SUM", col('OPEN_PRICE')+col('HIGH_PRICE')+col('LOW_PRICE')+col('CLOSE_PRICE'))
```

```python
df1.show(vertical = True)
```

```python
df1 = df1.filter(col('ALL_SUM')>col('NET_TRDVAL'))
```

```python
df1.select('NET_TRDVAL', 'ALL_SUM').show()
```

```python
#6
highest_net_trade_quantity = df.groupBy(['series']).max('NET_TRDQTY') 
```

```python
highest_net_trade_quantity.show()
```

```python
highest_net_trade_quantity = highest_net_trade_quantity.sort(desc("max(NET_TRDQTY)"))
```

```python
highest_net_trade_quantity.first()
```

```python
#7
high_low_open = spark.sql('SELECT max(OPEN_PRICE), min(OPEN_PRICE) from sharemarket')
```

```python
#8
trades_80 = spark.sql("select series, trades from sharemarket where trades>80")
```

```python
trades_80.show()
```

```python
#9
diff_net_trade = spark.sql("select series, sum(NET_TRDVAL) - sum(NET_TRDQTY) as Difference from sharemarket group by series")
```

```python
diff_net_trade.show()
```
